# 🚀 Automated ETL Pipeline for Data Cleaning & Reporting

Este proyecto nace de una necesidad crítica en las áreas de analítica y operaciones: el tiempo que se pierde limpiando datos a mano antes de poder generar informes ejecutivos.

Para resolver este cuello de botella, desarrollé de forma autónoma este **Pipeline ETL en Python** que automatiza la ingesta de datos crudos (estructuras complejas como JSON), realiza una limpieza rigurosa empleando Pandas y exporta reportes visuales listos para la toma de decisiones.

---

## 🛠️ Arquitectura del Proyecto (4 Bloques Operativos)

El código a continuación fue diseñado bajo un enfoque **modular y reutilizable**, asegurando que el sistema sea escalable y fácil de adaptar:

1. **Arranque:** Bloque encargado de la parametrización inicial, definición de rutas y preparación del entorno de ejecución.
2. **Materia Prima:** Ingesta y lectura automatizada de los datos crudos desde el servidor.
3. **Transformación:** Fase donde se ejecuta la lógica de negocio; limpieza de datos, tipado correcto y filtrado de anomalías empleando **Pandas**.
4. **Reporte:** Exportación y generación automatizada de reportes visuales ejecutivos con **Matplotlib**.

---
### ⚙️ CÓDIGO FUENTE EN VIVO:

```python
#El Gran Proyecto Integrador!

import pandas as pd
import matplotlib.pyplot as plt
def calcular_impuesto(monto):
  if monto>500:
    return monto*0.05
  else:
    return monto*0.02

datos_servidor = [
    {"transaccion": "T1", "activo": "Acciones Tech", "monto_usd": 800},
    {"transaccion": "T2", "activo": "Divisas Forex", "monto_usd": 350},
    {"transaccion": "T3", "activo": "Acciones Tech", "monto_usd": 1200},
    {"transaccion": "T4", "activo": "Criptoactivos", "monto_usd": 150},
    {"transaccion": "T5", "activo": "Divisas Forex", "monto_usd": 600}
]

df_operaciones = pd.DataFrame(datos_servidor)

df_operaciones["Impuesto_USD"] = [calcular_impuesto(monto) for monto in df_operaciones["monto_usd"]]

df_resumen = df_operaciones.groupby("activo")["monto_usd"].sum().reset_index()

plt.figure(figsize=(7,4))
plt.bar(df_resumen["activo"],df_resumen["monto_usd"],color="teal")
plt.title("Variación de los montos entre los activos")
plt.xlabel("Activos")
plt.ylabel("Montos")
plt.grid(axis="y",linestyle="--",alpha=0.5)
plt.show()

```
